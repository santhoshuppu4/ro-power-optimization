```python
import pandas as pd
import numpy as np

def add_timestamp(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]

    if "timestamp" in df.columns:
        # already exists
        df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
    elif "Timestamp" in df.columns:
        df["timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")
    elif ("Date" in df.columns) and ("Time" in df.columns):
        df["timestamp"] = pd.to_datetime(
            df["Date"].astype(str) + " " + df["Time"].astype(str),
            errors="coerce"
        )
    else:
        raise ValueError(f"No recognizable time columns. Found: {list(df.columns)}")

    df = df.dropna(subset=["timestamp"]).sort_values("timestamp")
    df = df.drop_duplicates(subset=["timestamp"], keep="last")
    return df
```

```python
def coerce_numeric(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    for c in df.columns:
        if c not in ["Date", "Time", "Timestamp", "timestamp"]:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def basic_clean(df: pd.DataFrame) -> pd.DataFrame:
    df = add_timestamp(df)
    df = coerce_numeric(df)

    # Drop rows with missing target (power) for training
    if "RO Power Consumption (W)" in df.columns:
        df = df.dropna(subset=["RO Power Consumption (W)"])

    return df
```

```python
FLOW_COLS = [
    "RO Overall Concentrate Flowrate (gpm)",
    "RO Overall Permeate Flowrate (gpm)",
    "RO Recirculation Flow Rate (gpm)"
]
TARGET = "RO Power Consumption (W)"

def denoise_rolling(df: pd.DataFrame, cols, window_seconds=60, sample_seconds=3):
    df = df.sort_values("timestamp").copy()
    win = max(3, int(window_seconds / sample_seconds))

    for c in cols:
        if c in df.columns:
            med = df[c].rolling(win, min_periods=1, center=True).median()
            sm  = med.rolling(win, min_periods=1, center=True).mean()
            df[c + "__denoised"] = sm
    return df
#For 10-min data, use sample_seconds=600.
```

```python
GALLON_TO_M3 = 0.00378541

def total_volume_m3(df: pd.DataFrame, flow_col="RO Overall Permeate Flowrate (gpm)"):
    df = df.sort_values("timestamp").copy()
    dt_min = df["timestamp"].diff().dt.total_seconds().fillna(0) / 60.0
    gallons = (df[flow_col] * dt_min).sum()
    return gallons * GALLON_TO_M3

# Example:
# vol = total_volume_m3(constant_3s_clean)
# print("Total m3 in dataset:", vol)
#verified the time-integral of permeate flow equals the baseline volume requirement.
```
