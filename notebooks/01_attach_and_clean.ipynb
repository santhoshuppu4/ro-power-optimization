```python
import os
import pandas as pd
import numpy as np

DATA_DIR = r"C:\Users\uvvss\Downloads\ro-power-optimization\data\Data"
constant_10m = pd.read_excel(os.path.join(DATA_DIR, "Constant_Flow_Rates_10min.xlsx"))
varied_10m   = pd.read_excel(os.path.join(DATA_DIR, "Varied_Flow_Rates_10min.xlsx"))
```

```python
TARGET = "RO Power Consumption (W)"

def clean_10min(df):
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]

    df["timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")
    df = df.dropna(subset=["timestamp"]).sort_values("timestamp")

    for c in df.columns:
        if c not in ["Timestamp", "timestamp"]:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    df = df.dropna(subset=[TARGET])
    return df
```

```python
constant_10m_clean = clean_10min(constant_10m)
varied_10m_clean   = clean_10min(varied_10m)

print(constant_10m_clean.shape)
print(varied_10m_clean.shape)
```
